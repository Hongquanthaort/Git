bai 5

```python
import random
danhsach = []
check = 0
for i in range(1000):
    danhsach.append(random.randint(-1000, 1000))
filename = input("Tên file: ")
print(len(danhsach))
for i in danhsach:
    if check < 9:
        f = open(filename, 'a')
        chuoikytu = str(i)+','
        f.write(chuoikytu)
        f.close()
        check +=1
    else:
        f = open(filename, 'a')
        f.write(str(i)+"\n")
        f.close()
        check = 0
f = open(filename, 'r')
a = f.readlines()
for i in a:
    print(i.replace(",", "    "))
f.close()
```
